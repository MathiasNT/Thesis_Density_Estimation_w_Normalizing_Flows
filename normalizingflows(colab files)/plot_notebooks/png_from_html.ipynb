```
import os
import time
from selenium import webdriver
from pathlib import Path



delay=5
plot_name = 'nyc_combi_large_jfc'
plot_folder = f'{root_path}/plot_notebooks/generated_plots'
driver_path = f'{root_path}/plot_notebooks/geckodriver.exe'
tmpurl = f'{plot_folder}/{plot_name}.html'
html_file = Path.cwd() / tmpurl

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("--window-size=845,920")
browser = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
#browser = webdriver.Chrome('chromedriver')
browser.get(html_file.as_uri())
#Give the map tiles some time to load
time.sleep(delay)
browser.save_screenshot(f'{plot_folder}/{plot_name}.png')
browser.quit()
```
